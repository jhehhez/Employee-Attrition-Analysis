```python
missing_values_per_column = df.isnull().sum()
print("Missing values per column:\n", missing_values_per_column)

total_missing_values = missing_values_per_column.sum()
print("\nTotal missing values:", total_missing_values)

features_have_missing_values = missing_values_per_column[missing_values_per_column > 0]
print("\nFeatures have missing values:\n", features_have_missing_values)
```

```python
feature_to_drop = missing_values_per_column.idxmax()
print(feature_to_drop)
```

```python
df = df.drop(columns=feature_to_drop)
```

```python
df = df.fillna(df.select_dtypes(include='number').mean())
```

```python
total_missing_values = df.isnull().sum().sum()
print("Total missing values:", total_missing_values)
```

```python
df['left'] = df['left'].map({'yes': 1, 'no': 0})
```

```python
print(df['left'].head())
```

```python
df.to_csv('cleaneddata.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_uniform, test_uniform = train_test_split(df, test_size=0.2, random_state=42)
```

```python
print("Uniform Sampling:")
print(f"Training set size: {len(train_uniform)}")
print(f"Test set size: {len(test_uniform)}")
print("Class distribution in test set:\n", test_uniform['left'].value_counts())
```

```python
train_stratified, test_stratified = train_test_split(df, test_size=0.2, random_state=42, stratify=df['left'])
```

```python
print("Stratified Sampling:")
print(f"Training set size: {len(train_stratified)}")
print(f"Test set size: {len(test_stratified)}")
print("Class distribution in test set:\n", test_stratified['left'].value_counts())
```
